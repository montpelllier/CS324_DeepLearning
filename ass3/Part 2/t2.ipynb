```python
from PIL import Image

def show_images(num_list):
    total_num = 938 * 200
    for num in num_list:
        path = './gan_images/{}.png'.format(num)
        print("Train {:.3f}% :".format(num/total_num*100))
        display(Image.open(path))
```

```python
start_list = [0, 1500, 3000, 4500, 6000]
print('The start of training:', start_list)

show_images(start_list)
```
